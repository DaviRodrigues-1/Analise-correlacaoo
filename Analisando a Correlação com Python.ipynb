```python
import pandas as pd
import seaborn as sns   
import matplotlib.pyplot as plt 
import numpy as np  
```

```python
df = pd.read_csv('data_tratada_MOD08.csv')
# Calculando a correlao
df.select_dtypes(include=['number']).corr() 

# Plotando correlao com seaborn, usando um heatmap 
heat_data = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(16,8))
sns.heatmap(heat_data, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size":10})
plt.title('Mapa de Correlao entre Variveis Numricas', fontsize=16)
plt.show()
```
